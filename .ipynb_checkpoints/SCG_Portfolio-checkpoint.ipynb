```python
## Import the relevant packages required
import ast
import pandas as pd
import re
import requests
import time
import urllib
import seaborn as sb
from requests.exceptions import RequestException
from bs4 import BeautifulSoup
```

```python
## Locate the relevant Scentre locations

url = "https://www.westfield.com.au/"

req = urllib.request.Request(
    url, 
    data=None, 
    headers={
        'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'
    }
)

response = urllib.request.urlopen(req)

html = response.read().decode('utf-8')

tags = re.findall('{(.*)}', html)

location_tags = tags[4]
```

```python
testing = location_tags

locations_dict = []

testing = testing.split('"ROOT_QUERY')
for each in testing:
    if "title" in each and len(each)<400:
        locations_dict += [each]
```

```python
locations_dict
i = 0
for each in locations_dict:    
    try:
        found = re.search('{"title"(.+?)}', each).group(1)
        string = '{"title"'+found+'}'
        locations_dict[i] = ast.literal_eval(string)
    except AttributeError:
        found = '' #just do nothing here
    i += 1
```

```python
## Convert locations to dictionaries to be called later.
locations_dict = pd.DataFrame(locations_dict)
locations_dict
```

```python
stores_df = locations_dict.loc[:,['state','slug']]
```

```python
## Include a user_agent for authentication.
user_agent = 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_9_3) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/35.0.1916.47 Safari/537.36'

i = 0

for each in stores_df['slug']:
    url = "https://www.westfield.com.au/" + each + "/browse-stores"
    
    response = requests.get(url,headers={'User-Agent': user_agent})

    ## Parse the response with BS.
    soup = BeautifulSoup(response.text,'html')

    all_shops = []
    for each in soup.findAll('a'):
        if 'index__headingLink___2YzqB' in each['class']:
            all_shops += [each.text]
            each = each.text
            if len(each.split('(Temporary closure)'))!=1:
                stores_df.loc[i,each.split('(Temporary closure)')[0].strip()] = 'Temporary closure'
            elif len(each.split('(Opening Soon)'))!=1:
                stores_df.loc[i,each.split('(Opening Soon)')[0].strip()] = 'Opening Soon'
            elif len(each.split('(Take away only)'))!=1:
                stores_df.loc[i,each.split('(Take away only)')[0].strip()] = 'Take away only'
            elif len(each.split('(Closed)'))!=1:
                stores_df.loc[i,each.split('(Closed)')[0].strip()] = 'Closed'
            else:
                stores_df.loc[i,each.strip()] = 'Normal'
    i += 1
        
```

```python
store_summary = pd.DataFrame(stores_df[['slug','state']])
store_summary['Total Stores'] = 0

states = stores_df['state'].unique()

modified_stores_df = stores_df
modified_stores_df = modified_stores_df.drop('state',axis=1)
modified_stores_df = modified_stores_df.drop('slug',axis=1)
modified_stores_df = modified_stores_df.transpose()

for each in range(len(modified_stores_df.columns)):
    store = modified_stores_df[each].value_counts()
    counter = 0
    try:
        counter += store['Normal']
        store_summary.loc[each,'Normal'] = store['Normal']
    except:
        True
    try:
        counter += store['Temporary closure']
        store_summary.loc[each,'Temporary closure'] = store['Temporary closure']
    except:
        True
    try:
        counter += store['Take away only']
        store_summary.loc[each,'Take away only'] = store['Take away only']
    except:
        True
    try:
        counter += store['Opening Soon']
        store_summary.loc[each,'Opening Soon'] = store['Opening Soon']
    except:
        True
    store_summary.loc[each,'Total Stores'] = counter

```
